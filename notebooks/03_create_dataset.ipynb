```python
import os

BASE_DIR = "../data/curated/drsprg/"
DATASET = os.path.join(BASE_DIR, "data_list_export_120217.xlsx")
OUTPUT_DATASET = DATASET.replace("curated", "processed")
```

```python
import pandas as pd
```

## Clean the data set

```python
columns_dict = {
    "AGE": "age",
    "STUDY NAME": "study_name",
    "GENDER": "gender",
    "WEIGHT": "weight",
    "HEIGHT": "height",
    "BMI": "bmi",
    "BSA - H": "bsa_h",
    "BSA - D": "bsa_d",
    "BPS1": "bps1",
    "BPD1": "bpd1",
    "BPD1.1": "bpd11",
    "BPS2": "bps2",
    "BPD2": "bpd2",
    "BPS1.1": "bps11",
    "BPS2.1": "bps21",
    "BPD2.1": "bpd21",
    "CKD": "ckd",
    "LK": "lk",
    "SCR": "scr",
    "CRC": "crc",
    "MAG3": "mag3",
    "EDTA": "edta",
    "FLOOD": "flood"
}
df = pd.read_excel(DATASET, header=0)
df.rename(columns=columns_dict, inplace=True)
df.drop(df.columns[10:], inplace=True, axis=1)
df.drop(0, inplace=True)
df.head()
```

```python
df.info()
```

```python
patients_to_remove = ["drsprg_025", "drsprg_055", "drsprg_065", "drsprg_081", "drsprg_106"]
for p in patients_to_remove:
    try:
        row_number = df[df["study_name"] == p].index.values[0]
        df.drop(row_number, axis=0, inplace=True)
    except:
        print(p)
```

```python
df.info()
```

```python
df.to_excel(OUTPUT_DATASET)
```
